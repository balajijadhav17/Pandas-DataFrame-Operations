# Using "numpy" Create a DataFrame

```python
import numpy as np
n=np.arange(1,82)
Roll_No=list(n)
Name = [
    "Aarav", "Vivaan", "Aditya", "Kabir", "Ishaan", "Dhruv", "Arjun", "Sai", "Krishna", "Rohan",
    "Siddharth", "Ananya", "Diya", "Ishika", "Meera", "Aarohi", "Aanya", "Myra", "Tanvi", "Sanjana",
    "Lakshmi", "Aryan", "Kartik", "Om", "Dev", "Gaurav", "Harsh", "Nikhil", "Abhishek", "Rahul",
    "Sneha", "Pooja", "Neha", "Komal", "Ritu", "Preeti", "Priya", "Anjali", "Kavya", "Radhika",
    "Vikram", "Ashwin", "Rajesh", "Ajay", "Manoj", "Suresh", "Ganesh", "Sumit", "Mohan", "Arvind",
    "Sahil", "Parth", "Aniket", "Yash", "Kunal", "Ritika", "Riya", "Simran", "Tanya", "Avni",
    "Ira", "Niharika", "Akash", "Jay", "Tejas", "Raghav", "Anmol", "Pranav", "Tarun", "Harsha",
    "Deepak", "Aakash", "Keshav", "Varun", "Bhavya", "Palak", "Shivani", "Mitali", "Pritam", "Chirag", 
    "Deeksha"
]

Division=["C"]*81
i1=np.random.randint(3,11,81)
ISE_1_Marks=list(i1)
i2=np.random.randint(6,11,81)
ISE_2_Marks=list(i2)
m=np.random.randint(12,31,81)
MSE_Marks=list(m)
e=np.random.randint(20,51,81)
ESE_Marks=list(e)
Grade=np.random.choice(["A+","A","B+","B","C+","C"],size=81,replace=True)
c=np.random.randint(2,5,81)
Creadit=list(c)
Branch=["CSE"]*81
Eng_Year=["Third_year"]*81


df=pd.DataFrame(
    {"Roll_No":Roll_No,"Name":Name,"Division":Division,
    "ISE_1_Marks":ISE_1_Marks,"ISE_2_Marks":ISE_2_Marks,"MSE__Marks":MSE_Marks,"ESE_Marks":ESE_Marks,"Grade":Grade,"Creadit":Creadit,"Branch":Branch,"Eng_Year":Eng_Year}
)
print(df)
```

# Perform Future Engineering  Apply Transformation 

```python
df["Total Marks"]=df["ISE_1_Marks"]+df["ISE_2_Marks"]+df["MSE__Marks"]+df["ESE_Marks"]
```

```python
df
```

# Student with less than 60 marks

```python
r = df[df["Total Marks"] < 60][["First_Name","Total Marks"]]
r
```

```python
# Student with more than 80 marks
```

```python
r1 = df[df["Total Marks"] > 80][["First_Name","Total Marks"]]
r1
```

# Student with the marks between(91,100)

```python
r2 = df[df["Total Marks"].between(91,100)][["First_Name","Total Marks"]]
r2
```

# Top Three student of the class according to their maximum total marks

```python
top_three=df.sort_values(by="Total Marks", ascending=False).head(3)
top_three
```

# Class Result Pass OR Fail 

```python
df["Result"]=df["Total Marks"].apply(lambda x: "Fail" if x < 65 else "Pass")
```

```python
df
```

```python
# Another way using numpy

import numpy as np
df["Performance"] = np.where(df["Total Marks"] > 85  , "Excellent", "Good")
```

```python
df
```

```python
ex=df[df["Performance"]=="Excellent"][["First_Name","Performance"]]
ex
```

# Pass Student Name

```python
passed_student=df[df["Result"]=="Pass"][["First_Name","Result"]]
passed_student
```

```python
failed_student=df[df["Result"]=="Fail"][["First_Name","Result","Grade"]]
failed_student
```
